```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Dataset tips
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.histplot(data=tips, x="total_bill", bins=20)
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.show()
```

```python
print(tips['total_bill'].corr(tips['tip']))
```

```python
sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="sex", # separa por cor para cada sexo
    scatter_kws={"alpha":0.6}, # transparência dos pontos
    line_kws={"linewidth":2}, # espessura da linha de tendência
)

plt.title("Relação entre valor da conta e gorjeta, por sexo")
plt.show()
```

```python
colunas_selecionadas = ['total_bill', 'tip', 'size']
corr = tips[colunas_selecionadas].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
import numpy as np
```

```python
data = np.array([2, 3, 5, 7, 8, 10, 11, 50])
print(np.mean(data))
```

```python
data = np.array([2, 3, 5, 7, 8, 10, 50])
print(f"Média: {np.mean(data)}")
```

```python
data = np.array([2, 3, 5, 7, 8, 10, 11, 50])
print(f"Mediana: {np.median(data)}")
```

```python
data = np.array([2, 3, 5, 7, 8, 10, 11])

Q1 = np.quantile(data, 0.25)
Q3 = np.quantile(data, 0.75)

print("Primeiro Quartil (Q1):", Q1)
print("Terceiro Quartil (Q3):", Q3)
```

```python
IRQ = Q3 - Q1
#Imprime o resultado do IRQ (Intervalo Interquartil)
print(IRQ)
```
